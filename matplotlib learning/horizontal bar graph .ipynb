```python
import matplotlib.pyplot as plt
import numpy as np

x=['python','C','C++','Java']
y=[1.9,2.7,1.8,3.9]
z=['Pro','Average','Noob','Fast']

width=0.2
p=np.arange(len(x))
p1= [j+width for j in p]

plt.xlabel("Languages",fontsize=18)
plt.ylabel("Number",fontsize=20)
plt.title("Happy happy happy")

c=['b','r','y','g']
plt.bar(p,y,width,color='r',label='Programs')
plt.bar(p1,z,width,color='g',label='Programs')

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```
